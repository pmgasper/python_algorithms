Graph common big O notations

```python
%matplotlib inline
import math
import matplotlib.pyplot as plt

n_list = range(1, 100)

o_one = [1 for n in n_list]
o_log_n = [math.log(n, 2) for n in n_list]
o_sqrt_n = [math.sqrt(n) for n in n_list]
o_n = [n for n in n_list]
o_n_log_n = [n * math.log(n, 2) for n in n_list]
o_n_sqrd = [n ** 2 for n in n_list]
o_2_n = [2 ** n for n in n_list]
o_n_fac = [math.factorial(n) for n in n_list]
```

```python
plt.figure(figsize=(8, 8))
plt.plot(n_list, o_one, label='O (1)')
plt.plot(n_list, o_log_n, label='O (log(n))')
plt.plot(n_list, o_sqrt_n, label='O (sqrt(n))')
plt.plot(n_list, o_n, label='O (n)')
plt.plot(n_list, o_n_log_n, label='O (n log(n))')
plt.plot(n_list, o_n_sqrd, label='O (n^2)')
plt.plot(n_list, o_2_n, label='O (2^n)')
plt.plot(n_list, o_n_fac, label='O (n!)')


plt.legend()
plt.xlabel('n')
plt.ylabel('N')
plt.axis((0, 100, 0 , 100))
```
